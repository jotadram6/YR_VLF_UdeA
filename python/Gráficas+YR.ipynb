```python
import matplotlib.pyplot as plt
import numpy as np
import random 

# RANGO EN X
x = np.arange(0.0, 3, 0.01)

# VALOR MEDIO Y ERROR

sigma =2
y=1-x*x

# CURVAS DE ERROR
onesigma=y+sigma
twosigma=y+sigma+sigma
menussigma=y-sigma
menustwosigma=y-sigma-sigma

#LABELS
plt.title(r'CMS $H\to\tau\tau \ \sqrt{s}=14 \ TeV,L=3000fb^{-1}$')
plt.ylabel(r'95% CL limit on $\sigma/\sigma_{SM}$')
plt.xlabel(r'$m_H(GeV)$')


plt.plot(x,y,color='black',label='Expected',linestyle=':')
plt.plot(x,onesigma,color='lawngreen')
plt.plot(x,twosigma,color='yellow')
plt.plot(x,menussigma,color='lawngreen')
plt.plot(x,menustwosigma,color='yellow')

#LLENADO ENTRE LINEAS



plt.fill_between(x, y, onesigma,color='lime',label='Expected (68%)')
plt.fill_between(x, y, menussigma,color='lime')

plt.fill_between(x, onesigma, twosigma,color='yellow',label='Expected (95%)')
plt.fill_between(x, menussigma, menustwosigma,color='yellow')

plt.legend()
plt.show()
```
